```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
path = "data.csv"
```

```python
df = pd.read_csv(path , skiprows = 4)
```

```python
# Select country name, country code, indicator, and population of a specific year 
year = "2020"

data1 = df[[
    "Country Name",
    "Country Code",
    "Indicator Name",
    year ]].dropna()

data1.head()
```

```python
top10 = data1.sort_values(by = year , ascending = True ).head(10)
```

```python
plt.figure()
plt.bar(top10["Country Code"], top10[year])
plt.xlabel("Country Code")
plt.ylabel("Population")
plt.title("least 10 Countries by Population (2020)")
plt.xticks(rotation=90, ha="right")
plt.tight_layout()
plt.show()
```
